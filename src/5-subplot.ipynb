### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the
# 
#  third, and a pie chart in the fourth.


import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = x ** 2  # Compute y values as square of x

# Line plot in first subplot
axes[0, 0].plot(x, y, label='y = x^2', color='b', linewidth=2)
axes[0, 0].set_title('Square Function Plot')
axes[0, 0].set_xlabel('X Values')
axes[0, 0].set_ylabel('Y Values (x^2)')
axes[0, 0].grid(True, linestyle='--', alpha=0.6)
axes[0, 0].legend()

# Bar plot in second subplot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'purple']
axes[0, 1].bar(categories, values, color=colors)
axes[0, 1].set_title('Category Values Bar Plot')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')

# Scatter plot in third subplot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='blue')
for i, txt in enumerate(zip(x_scatter, y_scatter)):
    axes[1, 0].annotate(f'{txt}', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='right')
axes[1, 0].set_title('Scatter Plot with Annotations')
axes[1, 0].set_xlabel('X Values')
axes[1, 0].set_ylabel('Y Values')
axes[1, 0].grid(True, linestyle='--', alpha=0.6)

# Pie chart in fourth subplot
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
axes[1, 1].set_title('Programming Language Popularity')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
